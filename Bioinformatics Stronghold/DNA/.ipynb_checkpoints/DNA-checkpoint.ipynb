```python
# Read dataset
with open('rosalind_dna.txt') as f:
    data = f.read()
```

```python
def count_DNA(seq: str):
#     print(f'A: {seq.count("A")}\nT: {seq.count("T")}\nG: {seq.count("G")}\nC: {seq.count("C")}')
    return(seq.count("A"), seq.count("T"), seq.count("G"), seq.count("C"))

A,T,G,C = count_DNA(data)

print(f'A: {A}\nT: {T}\nG: {G}\nC: {C}')
```

```python
# With Biopython
from Bio.Seq import Seq

seq = Seq(data)
# Seq objects also work like strings

A,T,G,C = count_DNA(seq)

print(f'A: {A}\nT: {T}\nG: {G}\nC: {C}')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot([A,T,G,C])
plt.show()
```

```python
# Is 'count' fast enough?
from random import choice
def gen_DNA(length):
    """
    Generate random DNA seq
    """
    DNA=""
    for count in range(length):
        DNA+=choice("CGTA")
    return DNA
```

```python
huge_dna = gen_DNA(100000000)
```

```python
count_DNA(huge_dna)
# Yeah, it's fast!
```
